# Árvore de regressão - parte 1

A base que utilizaremos contem os dados dos imóveis da california.
Considerem como variável Target (a ser prevista) a coluna median_house_value.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

### 1. Certifique-se de que esta base está no formato adequado para o scikitlearn.

```python
# Carregando o banco de dados

df = pd.read_csv('housing.csv')
print(df.shape)
df.head()
```

```python
# Avaliando os dados

df.info()
```

```python
# Removendo os dados faltantes, pois são uma pequena parcela do dataframe

df.dropna(inplace=True)
df.shape
```

```python
df.isna().sum()
```

> Não há mais dados faltantes, mas há uma coluna no formato `object` (string do pandas).

```python
# Verificando a variável

df.ocean_proximity.unique()
```

```python
# Criando categorias dummy
# Somente a variável "ocean_proximity" será afetada

df2 = pd.get_dummies(df).copy()
```

```python
print(df2.shape)
df2.head()
```

### 2.  Visualize a matriz de correlação, e veja se aparentemente você encontra alguma variável com potencial preditivo interessante.

```python
# Uma vez que o dataframe já foi "tratado", pode-se gerar a matriz de correlação

correlation = df2.corr()
correlation
```

```python
# Uma vez que se tem muitas variáveis é interessante gerar a visualização gráfica

# Configurando o sistema de cores
cmap = sns.diverging_palette(h_neg=850, h_pos=250, as_cmap=True, sep=60, center='light')
# Gerando a figura
sns.clustermap(correlation, figsize=[15, 15], cmap=cmap, annot=True)
```

```python
# A matriz de correlação também pode ser plotada diretamente por um .heatmap()
# - Não há organização hierárquica  de grupos no heatmap

plt.figure(figsize=[15,15])
sns.heatmap(correlation, annot=True)
```

## Insights

Variável que aparenta possuir maior potencial preditivo é "median_income", como pode-se notar é a variável com correlação mais alta em relação a "median_house_value".

### 3. Separe os dados em validação e teste

```python
# Uma vez que a variável Target (a ser prevista) é a coluna "median_house_value".
# A variável target "y"

y = df2['median_house_value']
y.head()
```

```python
# As variáveis explicativas 

# Removendo a variável Target e também 'longitude' e 'latitude',
# uma vez que as duas últimas não afetam a variável Target

X = df2.drop(columns=['median_house_value', 'longitude', 'latitude']).copy()
print(X.shape)
X.head()
```

```python
# Separando a base de dados para teste e treino

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=100)
```

```python
# Avaliando o shape dos dados 

print(X_train.shape) # 70% dos dados 
print(X_test.shape) # 30% dos dados
print(y_train.shape)
print(y_test.shape)
```

### 4. Treine duas árvores, uma com profundidade máxima = 8, outra com profundidade máxima = 2.

```python
# Definindo as ávores e suas profundidades

tree1 = DecisionTreeRegressor(max_depth=2)
tree2 = DecisionTreeRegressor(max_depth=8)
```

```python
# O "fitting" da primeira árvore

tree1.fit(X_train, y_train)
```

```python
# Segunda árvore

tree2.fit(X_train, y_train)
```

### 5. Calcule o MSE de cada uma das árvores do item anterior para a base de treinamento e para a base de testes.

```python
# Para a primeira árvore
train_tree1 = mean_squared_error(y_train, tree1.predict(X_train))
test_tree1 = mean_squared_error(y_test, tree1.predict(X_test))

print('--> Treino para a árvore com profundidade 2, o MSE é: {:.2f}'.format(train_tree1))
print('--> Teste para a árvore com profundidade 2, o MSE é: {:.2f}'.format(test_tree1))

# Para a segunda árvore
train_tree2 = mean_squared_error(y_train, tree2.predict(X_train))
test_tree2 = mean_squared_error(y_test, tree2.predict(X_test))

print('--> Treino para a árvore com profundidade 8, o MSE é: {:.2f}'.format(train_tree2))
print('--> Teste para a árvore com profundidade 8, o MSE é: {:.2f}'.format(test_tree2))
```

> De acordo com os valores apresentados pelo MSE o modelo pode não estar performando bem, isso se deve ao fato de que a diferença entre os valores preditos e os reais é muito grande. E, como o MSE retorna o quadrado dessa diferença, é possível obter valores demasiadamente grandes dependendo da variável em análise.


```python
# Gerando um painel "elemento a elemento" do cálculo (real - predito)^2
# O númerador sem considerar o somatório e razão em MSE

predict = tree1.predict(X_test)
real = y_test

values = (real - predict)**2

# Plot
plt.figure(figsize=[13,5])
sns.scatterplot(values, alpha=.5, color='blue')
plt.title('Relação entre os valores (real - predito)^2')
```

Realmente há valores bem discrepantes quando se analisa a relação (real - predito)^2. Para finalizar a análise, pode-se obter o MSE fazendo:

```python
mse = values.sum()/real.shape[0]
print('--> O valor do MSE para os dados de teste é: {:.2f}'.format(mse))
```

```python
# Analisando os valores de "median_house_value"

plt.figure(figsize=[10,5])
ax = sns.histplot(data=df, x='median_house_value', kde=True, color='blue', alpha=.5)
ax.axvline(df.median_house_value.mean(), color='red', linestyle='--', label='mean')
ax.legend()
```

```python
# Avaliando o coeficientes de determinação das árvores

for_train_tree1 = tree1.score(X_train, y_train)
for_test_tree1 = tree1.score(X_test, y_test)

for_train_tree2 = tree2.score(X_train, y_train)
for_test_tree2 = tree2.score(X_test, y_test)

template_train = 'O R-quadrado para o treino da árvore com profundidade {0} é : {1:.2f}.'
template_test = 'O R-quadrado para o teste da árvore com profundidade {0} é : {1:.2f}.'

print(template_train.format(tree1.get_depth(), for_train_tree1))
print(template_test.format(tree1.get_depth(), for_test_tree1))
print(template_train.format(tree2.get_depth(), for_train_tree2))
print(template_test.format(tree2.get_depth(), for_test_tree2))
```

 > Tendo os valores do coeficiente de determinação, **R-quadrado**, é possível afirmar que a árvore cuja a `max_depth=8` tem um melhor desempenho. Uma vez que o modelo, no teste, é capaz de explicar 64% da variância dos dados.

### 6. Com base nos resultados do item anterior, qual árvore te parece mais adequada?

Com base nos valores obtidos a árvore com `max_depth=8` apresenta melhores resultados. Pois possui um menor valor de MSE e, para essa métrica, quanto menor o valor melhor.

Juntamente com os dados obtidos para `R-quadrado` em que a árvore de maior profundidade é capaz de explicar uma maior proporção da variância dos dados.
